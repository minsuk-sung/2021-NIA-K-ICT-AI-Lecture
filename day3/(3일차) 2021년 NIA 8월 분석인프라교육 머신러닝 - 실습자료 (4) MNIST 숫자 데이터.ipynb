## 데이터 불러오기
<hr style="height:3px;border:none;color:#5F71F7;background-color:#203E60;" />

```python
from sklearn.datasets import fetch_openml
data_mnist = fetch_openml('mnist_784')
```

## 데이터 전처리

## 피처 엔지니어링
<hr style="height:3px;border:none;color:#5F71F7;background-color:#203E60;" />

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
```

```python
data_mnist['data'] = mms.fit_transform(data_mnist['data'])
```

## 모델링
<hr style="height:3px;border:none;color:#5F71F7;background-color:#203E60;" />

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_mnist['data'], 
                                                    data_mnist['target'], 
                                                    test_size=0.2,
                                                    random_state=42)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
model = MLPClassifier(hidden_layer_sizes=(100,),
                      activation='relu',
                      solver='adam',
                      learning_rate_init=0.001, 
                      verbose=1, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

## 모델 평가
<hr style="height:3px;border:none;color:#5F71F7;background-color:#203E60;" />

```python
model.score(X_test, y_test)
```

## 결과 예측
<hr style="height:3px;border:none;color:#5F71F7;background-color:#203E60;" />

```python
model.predict(X_test)
```
